```python
print("LETRA B:")

dados2018['data_inversa'] = pd.to_datetime(dados2018['data_inversa'])

#A UTILIZAÇÃO DO DATETIME() FOI NECESSÁRIA PARA TRANSFORMAR O TIPO DA COLUNA DATA_INVERSA EM UM DATATIME, FACILITANDO AS OPERAÇÕES EM DIANTE

porcentagem2018 = pd.DataFrame(dados2018.groupby(dados2018['data_inversa'].dt.month)['ilesos'].sum())

#NO CÓDIGO ACIMA, HOUVE A CRIAÇÃO DE UM DATAFRAME QUE MOSTRA A SOMA DE TODAS AS PESSOAS QUE SAÍRAM ILESAS EM CADA MÊS, SENDO NECESSÁRIA A UTILIZAÇÃO DA FUNÇÃO GROUPBY() PARA ARMAZENAR AS INFORMAÇÕES DE CADA MÊS, E A FUNÇÃO SUM() PARA SOMÁ-LAS

porcentagem2018['feridos graves'] = dados2018.groupby(dados2018['data_inversa'].dt.month)['feridos_graves'].sum()

#SEGUINDO A MESMA LÓGICA NA LINHA DE CÓDIGO ACIMA, FOI CRIADA A COLUNA FERIDOS GRAVES , QUE MOSTRA A SOMA DAS PESSOAS ENVOLVIDAS EM ACIDENTES GRAVES POR MÊS

porcentagem2018['total de pessoas'] = dados2018.groupby(dados2018['data_inversa'].dt.month)['pessoas'].sum()

#SEGUINDO A MEMSA LÓGICA, A COLUNA TOTAL DE PESSOAS MOSTRA O TOTAL DE PESSOAS QUE SE ENVOLVERAM EM QUALQUER ACIDENTE A CADA MÊS

porcentagem2018['frequência de pessoas ilesas'] = porcentagem2018['ilesos']/porcentagem2018['total de pessoas']

#O CÓDIGO ACIMA FOI RESPONSÁVEL POR CRIAR A VARIÁVEL FREQUÊNCIA DE PESSOAS ILESAS, QUE MOSTRA A RAZÃO DE PESSOAS ILESAS PELO O TOTAL DE PESSOAS

porcentagem2018['frequência de pessoas envolvidas em acidentes graves'] = (porcentagem2018['feridos graves']/porcentagem2018['total de pessoas'])

#SEGUINDO A MESMA LÓGICA, A COLUNA CRIADA ACIMA MOSTRA A RAZÃO DE FERIDOS GRAVES PELO O TOTAL DE PESSOAS ENVOLVIDAS EM ACIDENTES

#A MESMA LÓGICA UTILIZADA PARA CRIAR TODO O DATAFRAME ACIMA, SERÁ UTILIZADA NOVAMENTE PARA CRIAR UM DATAFRAME PARA 2019 E OUTRO PARA 2020
#SEGUE AGORA A CODIFICAÇÃO PARA CRIAR OS DATAFRAMES DE 2019 E 2020


#2019
dados2019['data_inversa'] = pd.to_datetime(dados2019['data_inversa'])
porcentagem2019 = pd.DataFrame(dados2019.groupby(dados2019['data_inversa'].dt.month)['ilesos'].sum())
porcentagem2019['feridos graves'] = dados2019.groupby(dados2019['data_inversa'].dt.month)['feridos_graves'].sum()
porcentagem2019['total de pessoas'] = dados2019.groupby(dados2019['data_inversa'].dt.month)['pessoas'].sum()
porcentagem2019['frequência de pessoas ilesas'] = porcentagem2019['ilesos']/porcentagem2019['total de pessoas']
porcentagem2019['frequência de pessoas envolvidas em acidentes graves'] = (porcentagem2019['feridos graves']/porcentagem2019['total de pessoas'])


#2020
dados2020['data_inversa'] = pd.to_datetime(dados2020['data_inversa'])
porcentagem2020 = pd.DataFrame(dados2020.groupby(dados2020['data_inversa'].dt.month)['ilesos'].sum())
porcentagem2020['feridos graves'] = dados2020.groupby(dados2020['data_inversa'].dt.month)['feridos_graves'].sum()
porcentagem2020['total de pessoas'] = dados2020.groupby(dados2020['data_inversa'].dt.month)['pessoas'].sum()
porcentagem2020['frequência de pessoas ilesas'] = porcentagem2020['ilesos']/porcentagem2020['total de pessoas']
porcentagem2020['frequência de pessoas envolvidas em acidentes graves'] = (porcentagem2020['feridos graves']/porcentagem2020['total de pessoas'])

#PARA MELHOR VISUALIZAÇÃO DOS DATAFRAMES, UTILIZAREI UMA CÉLULA DE CÓDIGO PARA CADA DATAFRAME GERADO
#COMEÇANDO PELO O DE 2018
print("ANO: 2018")
porcentagem2018
```

```python
print("LETRA C:")

df2020_copia = dados2020

#ACIMA EU CRIEI UMA CÓPIA DO DATAFRAME DO ANO DE 2020 ORIGINAL PARA QUE NÃO HOUVESSE MUDANÇAS NO ARQUIVO ORIGINAL

df2020_copia['data_inversa'] = pd.to_datetime(df2020_copia['data_inversa'])

#ACIMA EU TRANSFORMEI A COLUNA DE DATA_INVERSA DA CÓPIA PARA O TIPO DATATIME, PARA FACILITAR AS OPERAÇÕES POSTERIORES

df2020_copia['meses'] = df2020_copia['data_inversa'].dt.month

#NA LINHA DE CÓDIGO ANTERIOR, FOI CRIADO UMA COLUNA CHAMADA MESES COM A FUNÇÃO DT.MONTH PARA SEPARAR OS MESES DAS DATAS EM NÚMERO INTEIROS EQUIVALENTES AO MÊS

tamanho_2020 = (df2020_copia['meses']<= 2).sum()

#LOGO ACIMA, FOI FEITO A CONDIÇÃO PARA SOMAR A QUANTIDADE DE ACIDENTES DE 2020 SEM A PANDEMIA

len_18_19_20 = (len(dados2018) + len(dados2019) + tamanho_2020)/26

#ENQUANTO AQUI, FOI FEITA A CRIAÇÃO DA VARIÁVEL LEN_18_19_20 COM O OBJETIVO DE ARMAZENAR A MÉDIA DE ACIDENTES POR MÊS DOS ANOS 2018, 2019 E 2020 SEM A PANDEMIA

len_pandemia = ((df2020_copia['meses']>=3).sum())/10

#ENQUANTO O CÓDIGO A CIMA FOI FEITO A CONDIÇÃO DE SOMA PARA OS MESES DE 2020 QUE HOUVE A PANDEMIA


df_analise = pd.DataFrame(data=(len_18_19_20, len_pandemia), columns=['Média de acidente mensal'], index=[['Sem pandemia', 'Com pandemia']])

#ACIMA HOUVE A CRIAÇÃO DE UM DATAFRAME PARA FACILITAR A VISUALIZAÇÃO DOS DADOS

df_analise
```
